### Python Wrappers

```python
my_api_key = 'f72035084b9043abb7d52225212212'
base_url = 'https://api.weatherapi.com/v1'
api_method = '/current.json'
city = 'Chicago'

api_url = f'https://api.weatherapi.com/v1/current.json?key={my_api_key}&q=Chicago&lang=es'
api_url = f"{base_url}{api_method}?key={my_api_key}&q={city}"
print(api_url)
```

```python
res.json()
```

```python
class WeatherAPI:
    
    def __init__(self, api_key):
        self.base_url = 'https://api.weatherapi.com/v1'
        self.api_key = api_key
        
        
    def _get(self, q='', api_method):
        res = requests.get(f'{self.base_url}{api_method}?key={self.api_key}&q={q}')
        if res.ok:
            return res.json()
        
    def get_city_weather(self, city):
        data = self._get(city)
        display_name = f"{data['location']['name']}, {data['location']['region']}"
        current_temp = data['current']['temp_f']
        current_condition = data['current']['condition']['text']
        feels_like = data['current']['feelslike_f']
        degree_sign = u"\N{DEGREE SIGN}"
        print(f"It is currently {current_condition} and {current_temp}{degree_sign}F in {display_name}. It feels like {feels_like}{degree_sign}F")
        
        
    def get_forecast(self, city, days=3):
        forecast_data = self._get(city, 'forecast.json', days=days)
        display_name = f"{data['location']['name']}, {data['location']['region']}"
        list_of_days = forecast_data['forecast']['forecastday']
        print(f"Forecast for {display_name}:")
        for day in list_of_days:
            date = day['date']
            high = day['day']['maxtemp_f']
            low = day['day']['mintemp_f']
            condition = day['day']['condition']['text']
            degree_sign = u"\N{DEGREE SIGN}"
            print(f"\tOn {date}, it will be {condition} with a high of {high}{degree_sign}F and a low of {low}{degree_sign}F")

weather = WeatherAPI(my_api_key)
```

```python
weather.get_city_weather('Chicago')
```

Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
poke_base_url = 'https://pokeapi.co/api/v2/'
endpoint = 'pokemon/'


poke_url = f"{base_url}{endpoint}"
print(poke_url)

class PokemonAPI:
    def __init__(self):
        self.base_url = 'https://pokeapi.co/api/v2/'
           
    def _get(self, pokemon):
        res = requests.get(f"{base_url}{endpoint}{pokemon}")
        if res.ok:
            return res.json()

    def get_pokevitals(self, pokemon):
        data = self._get(pokemon)
        pokeheight = data['height']
        pokeweight = data['weight']
        print(f"{pokemon.title()} has a height of {pokeheight} and weight of {pokeweight}")
        
fetch = PokemonAPI()
```
